# 01_data_understanding.ipynb

This notebook focuses on understanding the raw dataset, including data loading, initial inspection, handling missing values and duplicates, identifying outliers, performing Exploratory Data Analysis (EDA), and engineering new features.

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('battery_cycle_level_dataset_CLEAN_FINAL.csv')

print('Dataset loaded successfully.')
```

## Initial Data Inspection

Let's view the first few rows and get a summary of the DataFrame to understand its structure and data types.

```python
# Display the first 5 rows of the DataFrame
display(df.head())

# Display information about the DataFrame, including data types and non-null values
df.info()
```

## Data Cleaning: Missing Values and Duplicates

Now, let's check for any missing values and duplicate rows in the dataset, which are crucial steps in data cleaning.

```python
# Check for missing values
missing_values = df.isnull().sum()
print('Missing Values:\n', missing_values[missing_values > 0])

# Check for duplicate rows
duplicate_rows = df.duplicated().sum()
print(f'\nNumber of Duplicate Rows: {duplicate_rows}')

# If there are duplicates, remove them
if duplicate_rows > 0:
    df.drop_duplicates(inplace=True)
    print(f'Duplicate rows removed. New shape: {df.shape}')
```

## Data Cleaning: Outlier Detection and Handling

Based on the problem description, some values in `voltage`, `temperature`, and `soh` columns can be physically impossible. Let's define reasonable ranges and filter out any data points that fall outside these ranges.

### Plausible Ranges:
- **Voltage**: A typical battery voltage range might be around 2.5V to 4.2V. We'll use a slightly wider range to be safe.
- **Temperature**: Human-detectable temperatures are generally above -20°C and below 100°C for battery operation.
- **SOH**: State of Health is a percentage, so it should be between 0% and 100% (inclusive).

```python
# Define plausible ranges based on general battery characteristics and physical limits
voltage_min, voltage_max = 2.0, 4.5 # Volts
temperature_min, temperature_max = -20.0, 100.0 # Celsius
soh_min, soh_max = 0.0, 1.0 # As a proportion (0% to 100%)

# Identify outliers
outliers_before = df[
    (df['voltage'] < voltage_min) | (df['voltage'] > voltage_max) |
    (df['temperature'] < temperature_min) | (df['temperature'] > temperature_max) |
    (df['soh'] < soh_min) | (df['soh'] > soh_max)
]

print(f"Number of rows with identified outliers: {len(outliers_before)}")

if not outliers_before.empty:
    print("\nExamples of identified outliers:")
    display(outliers_before.head())

    # Remove outliers
    df_cleaned = df[
        (df['voltage'] >= voltage_min) & (df['voltage'] <= voltage_max) &
        (df['temperature'] >= temperature_min) & (df['temperature'] <= temperature_max) &
        (df['soh'] >= soh_min) & (df['soh'] <= soh_max)
    ].copy()

    print(f"\nShape before outlier removal: {df.shape}")
    print(f"Shape after outlier removal: {df_cleaned.shape}")
    df = df_cleaned # Update the DataFrame
else:
    print("No outliers found within the defined ranges. Proceeding with the original DataFrame.")

# Verify SOH is within 0-1 range after cleaning (if soh is represented as 0-100, normalize it)
if df['soh'].max() > 1.05: # Check if soh values are likely in 0-100 range
    print("\nSOH values appear to be in 0-100 range. Normalizing to 0-1.")
    df['soh'] = df['soh'] / 100.0
    print(f"SOH range after normalization: {df['soh'].min():.2f} - {df['soh'].max():.2f}")
```

## Phase 3: Feature Engineering

Now that we've cleaned the data and performed initial EDA, it's time to create new, more expressive features that can help the XGBoost model capture complex relationships in the data. As outlined, we will derive several new features from the existing ones.

```python
# It's important to calculate 'max_capacity' per battery_id, if relevant for capacity loss.
# For simplicity, we'll assume a global max_capacity for the initial calculation if not specified.
# If `max_capacity` is dynamic per battery, a more complex approach would be needed.
# For now, let's use the maximum capacity observed in the entire dataset or initial capacity if available
# Let's assume initial capacity for each battery is its max capacity or highest observed.
# For simplicity, we can use the maximum capacity from the initial cycle for each battery_id

# Calculate max_capacity for each battery_id (assuming it's the capacity at the first cycle or max capacity overall)
max_capacities = df.groupby('battery_id')['capacity'].max().rename('max_capacity')
df = df.merge(max_capacities, on='battery_id', how='left')

# 1. Capacity Loss: 1 - (capacity / max_capacity)
df['capacity_loss'] = 1 - (df['capacity'] / df['max_capacity'])

# 2. Voltage × Temperature
df['voltage_x_temperature'] = df['voltage'] * df['temperature']

# 3. Temperature / Cycle (handle division by zero if cycle starts from 0, or add 1)
df['temperature_div_cycle'] = df['temperature'] / (df['cycle'] + 1) # Add 1 to avoid division by zero

# 4. Capacity / Cycle (handle division by zero if cycle starts from 0, or add 1)
df['capacity_div_cycle'] = df['capacity'] / (df['cycle'] + 1) # Add 1 to avoid division by zero

# 5. Voltage²
df['voltage_sq'] = df['voltage'] ** 2

# 6. Temperature²
df['temperature_sq'] = df['temperature'] ** 2

# 7. Cycle²
df['cycle_sq'] = df['cycle'] ** 2

print("Engineered features added successfully.")
print("New DataFrame head with engineered features:")
display(df.head())

# Drop the temporary max_capacity column as it's no longer needed after feature engineering
df = df.drop(columns=['max_capacity'])

print("New DataFrame info:")
df.info()
```

# 02_train_xgboost.ipynb

This notebook focuses on preparing the data for model training, performing a strategic train/test split, training an XGBoost Regressor model, evaluating its performance, and understanding feature importance.

## Phase 6: Train/Test Split

To ensure a robust and realistic evaluation of our model, we will not perform a random split. Instead, we'll split the data based on `battery_id`. This means certain batteries will be entirely in the training set, and others entirely in the test set. This approach prevents data leakage and provides a more accurate measure of the model's ability to generalize to new, unseen batteries.

```python
from sklearn.model_selection import train_test_split
import numpy as np

# Identify unique battery_ids
unique_battery_ids = df['battery_id'].unique().tolist()

# Split battery_ids into training and testing sets
# Using a fixed random_state for reproducibility
train_battery_ids, test_battery_ids = train_test_split(unique_battery_ids, test_size=0.2, random_state=42)

print(f"Total unique battery_ids: {len(unique_battery_ids)}")
print(f"Training battery_ids: {len(train_battery_ids)}")
print(f"Testing battery_ids: {len(test_battery_ids)}")

# Create training and testing DataFrames based on the split battery_ids
df_train = df[df['battery_id'].isin(train_battery_ids)].copy()
df_test = df[df['battery_id'].isin(test_battery_ids)].copy()

print(f"\nShape of training data: {df_train.shape}")
print(f"Shape of testing data: {df_test.shape}")

# Define features (X) and target (y)
# 'battery_id', 'soh', 'rul' and 'max_capacity' (if it wasn't dropped) are not features for training
features = [col for col in df.columns if col not in ['battery_id', 'soh', 'rul']]

X_train = df_train[features]
y_train = df_train['soh']
X_test = df_test[features]
y_test = df_test['soh']

print("\nTrain/Test split completed successfully based on battery_id.")
print(f"Number of features: {len(features)}")
print(f"Features used: {features}")
```

## Phase 7: Feature Scaling?

As noted in the problem description, tree-based models like XGBoost generally do not require feature scaling (normalization or standardization). This is because decision trees are not affected by the scale of the features, as they make decisions based on thresholds that split the data. Therefore, we will skip this step and proceed directly to model training.

## Phase 8: Train XGBoost

Now, we will train an XGBoost Regressor model using our prepared training data. We'll use default parameters initially and can tune them later if needed.

```python
import xgboost as xgb

# Initialize the XGBoost Regressor model
# Using some common parameters for regression
xgb_model = xgb.XGBRegressor(
    objective='reg:squarederror', # Objective for regression tasks
    n_estimators=100,             # Number of boosting rounds
    learning_rate=0.1,            # Step size shrinkage to prevent overfitting
    max_depth=5,                  # Maximum depth of a tree
    subsample=0.8,                # Subsample ratio of the training instance
    colsample_bytree=0.8,         # Subsample ratio of columns when constructing each tree
    random_state=42,              # Seed for reproducibility
    n_jobs=-1                     # Use all available CPU cores
)

print("XGBoost Regressor initialized. Training the model...")

# Train the model
xgb_model.fit(X_train, y_train)

print("XGBoost model training completed successfully.")
```

## Phase 9: Evaluate

After training the model, it's essential to evaluate its performance on the test set. We will use standard regression metrics to quantify how well our model predicts the State of Health (SOH).

### Regression Metrics:
- **Mean Absolute Error (MAE)**: The average of the absolute differences between predictions and actual values. It gives an idea of the average magnitude of errors.
- **Root Mean Squared Error (RMSE)**: The square root of the average of the squared differences between predictions and actual values. It penalizes larger errors more heavily.
- **R-squared (R² Score)**: Represents the proportion of the variance in the dependent variable that is predictable from the independent variables. A higher R² indicates a better fit.

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np # Import numpy for sqrt
import matplotlib.pyplot as plt
import seaborn as sns # Import seaborn for plotting

# Make predictions on the test set
y_pred = xgb_model.predict(X_test)

# Calculate evaluation metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred) # Calculate MSE first
rmse = np.sqrt(mse) # Then take the square root for RMSE
r2 = r2_score(y_test, y_pred)

print(f"Model Evaluation on Test Set:")
print(f"- Mean Absolute Error (MAE): {mae:.4f}")
print(f"- Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"- R-squared (R² Score): {r2:.4f}")

# Optionally, visualize predictions vs actual
plt.figure(figsize=(10, 6))
sns.scatterplot(x=y_test, y=y_pred, alpha=0.7)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2) # Ideal line
plt.xlabel('Actual SOH')
plt.ylabel('Predicted SOH')
plt.title('Actual vs Predicted SOH on Test Set')
plt.grid(True)
plt.show()
```

## Phase 10: Explain the Model - Feature Importance

Understanding which features contribute most to the model's predictions is crucial for interpretability and gaining domain insights. XGBoost models provide a way to access feature importance. We will visualize these importances to see which factors the model prioritized in estimating SOH.

```python
# Get feature importances from the trained XGBoost model
feature_importances = xgb_model.feature_importances_

# Create a pandas Series for better visualization
importance_df = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': feature_importances
}).sort_values(by='Importance', ascending=False)

print("Feature Importances:")
display(importance_df)

# Plot feature importances
plt.figure(figsize=(12, 8))
sns.barplot(x='Importance', y='Feature', data=importance_df, palette='viridis')
plt.title('XGBoost Feature Importance')
plt.xlabel('Importance (F-score)')
plt.ylabel('Feature')
plt.tight_layout()
plt.show()
```

### Phase 10 (Continued): Explain the Model - SHAP

While feature importance gives us a global understanding of which features are important, SHAP (SHapley Additive exPlanations) provides a way to explain individual predictions. This means we can understand how each feature contributed to a particular SOH prediction for a specific battery cycle.

SHAP values tell us how much each feature pushes the model's output from the base value (the average prediction).

```python
import shap

# Create a SHAP explainer object. For tree models like XGBoost, TreeExplainer is efficient.
explainer = shap.TreeExplainer(xgb_model)

# Calculate SHAP values for the test set
# This might take a moment depending on the size of X_test
shap_values = explainer.shap_values(X_test)

print("SHAP values calculated successfully. Let's visualize them.")

# Visualize the SHAP values for a single prediction (e.g., the first test sample)
print("\nSHAP explanation for the first test sample:")
shap.initjs()
shap.force_plot(explainer.expected_value, shap_values[0,:], X_test.iloc[0,:])
```

```python
# Plot a summary of the SHAP values to see overall feature impact and direction
print("\nSHAP summary plot (overall feature impact):")
shap.summary_plot(shap_values, X_test, plot_type="bar", max_display=len(features))
shap.summary_plot(shap_values, X_test, max_display=len(features))
```

## Phase 11: Export Model

To make our trained model usable for future predictions or deployment, we need to save it. We will export the `xgb_model` to a pickle file (`.pkl`).

```python
import joblib
import os

# Define the filename for the exported model
model_filename = 'battery_soh_model.pkl'

# Save the trained XGBoost model using joblib
joblib.dump(xgb_model, model_filename)

print(f"Model successfully exported as '{model_filename}'")
print(f"Model saved to: {os.path.abspath(model_filename)}")
```
